## Análise de dados

```python
from kaggle.api.kaggle_api_extended import KaggleApi
import pandas as pd
import os

# Inicializa e autentica a API
api = KaggleApi()
api.authenticate()

# Define o dataset e o diretório de destino
dataset = 'heptapod/titanic'
data_dir = 'titanic_data'
if not os.path.exists(data_dir):
    os.makedirs(data_dir)

# Baixa e descompacta automaticamente o arquivo 'train.csv'
api.dataset_download_file(dataset, 'train.csv', path=data_dir)

# Lê o arquivo CSV descompactado diretamente em um DataFrame pandas
df = pd.read_csv(os.path.join(data_dir, 'train.csv'))

# Exibe as primeiras linhas do DataFrame
print(df.head())
```
